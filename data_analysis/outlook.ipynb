```python
import pandas as pd
import numpy as np
import random
random.seed(42)
np.random.seed(42)
```

```python
#first function is enough
memory=pd.read_csv("/home/mlqin/qml/assignment/Distributed_System/Project1/data/app_memory_percentiles.anon.d01.csv")
duration=pd.read_csv("/home/mlqin/qml/assignment/Distributed_System/Project1/data/function_durations_percentiles.anon.d01.csv")
```

```python

memory_features=['HashOwner','HashApp','AverageAllocatedMb']
memory=memory[memory_features]
```

```python
duration_feature=['HashOwner','HashApp','HashFunction','Average']
duration=duration[duration_feature]
```

```python
duplicates = memory[memory.duplicated(subset=['HashOwner','HashApp'], keep=False)]
duplicates
```

```python
function_we_choose=duration['HashFunction'].unique()
```

```python
duplicate_function=pd.merge(duration, duplicates, on=['HashOwner', 'HashApp'], how='inner')['HashFunction'].unique()
all_function=duration['HashFunction'].unique()
wanted_mask = ~np.isin(all_function, duplicate_function)

# 使用布尔索引从full_array中提取你想要的元素
wanted_function = all_function[wanted_mask]
print(wanted_function)
```

```python
duration=duration[duration['HashFunction'].isin(wanted_function)]
merged_df = pd.merge(duration, memory, on=['HashOwner', 'HashApp'], how='left').drop_duplicates(subset=['HashOwner', 'HashApp'])
merged_df
```
